```python
# imports

import os, time, math, json, requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import yfinance as yf
from pandas_datareader import data as pdr

plt.rcParams["figure.figsize"] = (9,5)
```

```python
def load_multi_asset_data(start=None, end=None, years=5):
    """
    Returns a (prices_df, source_used) tuple.
    Columns: Equity_SPY, Rates_TLT, FX_UUP, Gold_GLD
    Priority: yfinance -> Stooq -> Synthetic
    """
    # 1) Try Yahoo Finance
    try:
        if start is None or end is None:
            prices = yf.download(["SPY","TLT","UUP","GLD"], period=f"{years}y", auto_adjust=True, progress=False)
        else:
            prices = yf.download(["SPY","TLT","UUP","GLD"], start=start, end=end, auto_adjust=True, progress=False)
        if isinstance(prices, pd.Series): prices = prices.to_frame()
        prices = prices["Close"] if "Close" in prices.columns else prices
        if prices.isna().all().all() or prices.empty:
            raise ValueError("Yahoo returned empty frame.")
        prices = prices.rename(columns={
            "SPY":"Equity_SPY", "TLT":"Rates_TLT", "UUP":"FX_UUP", "GLD":"Gold_GLD"
        }).dropna(how="all")
        return prices, "yahoo"
    except Exception as e:
        print("Yahoo failed, falling back to Stooq. Reason:", repr(e))

    # 2) Try Stooq via pandas-datareader (symbols end with .US)
    try:
        # Stooq is daily only, no intraday; no API key needed
        end_dt = datetime.today() if end is None else datetime.fromisoformat(end)
        start_dt = end_dt - timedelta(days=365*years) if start is None else datetime.fromisoformat(start)
        stooq_map = {"Equity_SPY":"SPY.US", "Rates_TLT":"TLT.US", "FX_UUP":"UUP.US", "Gold_GLD":"GLD.US"}
        frames = []
        for name, sym in stooq_map.items():
            df = pdr.DataReader(sym, "stooq", start_dt, end_dt)[["Close"]].rename(columns={"Close":name})
            frames.append(df)
        prices = pd.concat(frames, axis=1).sort_index().dropna(how="all")
        if prices.isna().all().all() or prices.empty:
            raise ValueError("Stooq returned empty frame.")
        return prices, "stooq"
    except Exception as e:
        print("Stooq failed, generating synthetic data. Reason:", repr(e))

    # 3) Synthetic GBM fallback (always works)
    np.random.seed(42)
    idx = pd.date_range(end=datetime.today(), periods=252*years, freq="B")
    specs = {
        "Equity_SPY": (0.08/252, 0.18/np.sqrt(252)),
        "Rates_TLT":  (0.03/252, 0.12/np.sqrt(252)),
        "FX_UUP":     (0.00/252, 0.06/np.sqrt(252)),
        "Gold_GLD":   (0.02/252, 0.15/np.sqrt(252)),
    }
    data = {}
    for k,(mu,sigma) in specs.items():
        steps = mu + sigma*np.random.randn(len(idx))
        data[k] = 100*np.exp(np.cumsum(steps))
    prices = pd.DataFrame(data, index=idx)
    return prices, "synthetic"


start = None  # or "2020-01-01"
end   = None
prices, source = load_multi_asset_data(start, end, years=5)
print("Data source used:", source)
prices.tail()
```

```python
# log returns
def log_ret(px): return np.log(px/px.shift(1)).dropna()
rets = log_ret(prices)

stats = pd.DataFrame({
    "ann_mean_log": rets.mean()*252,
    "ann_vol": rets.std()*np.sqrt(252),
})
stats.sort_values("ann_vol", ascending=False)
```

```python
# normalized plot and correlations
(prices/prices.iloc[0]*100).plot()
plt.title(f"Normalized Prices (source={source})"); plt.xlabel("Date"); plt.ylabel("Index=100")
plt.show()

corr = rets.corr(); corr
```

```python
# Coinbase (primary)
def coinbase_depth(product_id="BTC-USD", level=2):
    url = f"https://api.exchange.coinbase.com/products/{product_id}/book"
    r = requests.get(url, params={"level": level}, timeout=10, headers={"User-Agent":"quant-notebook"})
    r.raise_for_status()
    j = r.json()
    bids = np.array([[float(p), float(sz)] for p, sz, n in j["bids"]], dtype=float)
    asks = np.array([[float(p), float(sz)] for p, sz, n in j["asks"]], dtype=float)
    bids = bids[bids[:,0].argsort()[::-1]]
    asks = asks[asks[:,0].argsort()]
    best_bid, best_ask = bids[0,0], asks[0,0]
    spread = best_ask - best_bid
    return {"bids":bids, "asks":asks, "bid":best_bid, "ask":best_ask, "spread":spread, "src":"coinbase"}

# Kraken (first fallback)
def kraken_depth(pair="XXBTZUSD", count=500):
    url = "https://api.kraken.com/0/public/Depth"
    r = requests.get(url, params={"pair": pair, "count": count}, timeout=10)
    r.raise_for_status()
    j = r.json()
    key = list(j["result"].keys())[0]
    bids_raw = j["result"][key]["bids"]  # [price, volume, timestamp]
    asks_raw = j["result"][key]["asks"]
    bids = np.array([[float(p), float(v)] for p, v, ts in bids_raw], dtype=float)
    asks = np.array([[float(p), float(v)] for p, v, ts in asks_raw], dtype=float)
    bids = bids[bids[:,0].argsort()[::-1]]
    asks = asks[asks[:,0].argsort()]
    best_bid, best_ask = bids[0,0], asks[0,0]
    spread = best_ask - best_bid
    return {"bids":bids, "asks":asks, "bid":best_bid, "ask":best_ask, "spread":spread, "src":"kraken"}

# Bitstamp (second fallback)
def bitstamp_depth(pair="btcusd"):
    url = f"https://www.bitstamp.net/api/v2/order_book/{pair}/"
    r = requests.get(url, timeout=10)
    r.raise_for_status()
    j = r.json()
    bids = np.array([[float(p), float(q)] for p, q in j["bids"]], dtype=float)
    asks = np.array([[float(p), float(q)] for p, q in j["asks"]], dtype=float)
    bids = bids[bids[:,0].argsort()[::-1]]
    asks = asks[asks[:,0].argsort()]
    best_bid, best_ask = bids[0,0], asks[0,0]
    spread = best_ask - best_bid
    return {"bids":bids, "asks":asks, "bid":best_bid, "ask":best_ask, "spread":spread, "src":"bitstamp"}

# fetch with fallback
def get_order_book():
    # coinbase btcusdt first
    try:
        return coinbase_depth("BTC-USD", level=2)
    except Exception as e_cb:
        print("Coinbase failed:", repr(e_cb))
    # try kraken
    try:
        return kraken_depth("XXBTZUSD", count=500)
    except Exception as e_kr:
        print("Kraken failed:", repr(e_kr))
    # try bitstamp
    try:
        return bitstamp_depth("btcusd")
    except Exception as e_bs:
        print("Bitstamp failed:", repr(e_bs))
        raise RuntimeError("All public depth endpoints failed.")

# use it
book = get_order_book()
bids, asks = book["bids"], book["asks"]
bid, ask, spread, src = book["bid"], book["ask"], book["spread"], book["src"]
mid = (bid+ask)/2
print(f"Source: {src} | bid={bid:.2f} ask={ask:.2f} spread={spread:.2f}")
```

```python
# cumulative depth
cum_bid = np.cumsum(bids[:,1]); cum_ask = np.cumsum(asks[:,1])

plt.figure()
plt.plot(bids[:,0], cum_bid, label="Bid depth")
plt.plot(asks[:,0], cum_ask, label="Ask depth")
plt.title(f"BTC/USD Order Book Cumulative Depth ({src})")
plt.xlabel("Price"); plt.ylabel("Cumulative Size"); plt.legend(); plt.show()

# simple slippage for a buy market order
def simulate_market_buy(asks, qty):
    remaining, cost = qty, 0.0
    for price, size in asks:
        take = min(remaining, size)
        cost += take * price
        remaining -= take
        if remaining <= 0: break
    if remaining > 0:
        return np.nan
    return cost/qty

sizes = [0.01, 0.05, 0.10, 0.25, 0.50]
rows = []
for q in sizes:
    vwap = simulate_market_buy(asks, q)
    rows.append({"qty": q, "vwap": vwap, "slippage": (vwap - mid) if np.isfinite(vwap) else np.nan})
pd.DataFrame(rows)
```

```python
# intraday patterns

ALLOWED_GRANS = {60, 300, 900, 3600, 21600, 86400}  # per Coinbase
MAX_POINTS = 300  # max candles per request

def coinbase_candles_chunked(product_id="BTC-USD", granularity=60, start=None, end=None, pause_sec=0.2):
    """
    Fetch Coinbase OHLCV candles in chunks (<=300 points/request).
    Returns DataFrame: index=time (UTC), cols=[open, high, low, close, volume].
    """
    assert granularity in ALLOWED_GRANS, f"granularity must be one of {ALLOWED_GRANS}"
    if end is None:
        end = datetime.utcnow()
    if start is None:
        # default: last MAX_POINTS-1 bars (one chunk)
        start = end - timedelta(seconds=granularity*(MAX_POINTS-1))

    frames = []
    cursor_start = start

    while cursor_start < end:
        cursor_end = min(cursor_start + timedelta(seconds=granularity*(MAX_POINTS-1)), end)
        url = f"https://api.exchange.coinbase.com/products/{product_id}/candles"
        params = {
            "granularity": granularity,
            "start": cursor_start.isoformat(),
            "end": cursor_end.isoformat(),
        }
        # a slightly more generous timeout than 1s
        r = requests.get(url, params=params, timeout=10, headers={"User-Agent":"quant-notebook"})
        r.raise_for_status()
        data = r.json()

        # Coinbase returns list of [time, low, high, open, close, volume] DESC by time
        if isinstance(data, list) and len(data) > 0:
            df = pd.DataFrame(data, columns=["time","low","high","open","close","volume"])
            df["time"] = pd.to_datetime(df["time"], unit="s", utc=True)
            df = df.sort_values("time")
            frames.append(df)

        cursor_start = cursor_end
        # polite pacing to avoid rate limits
        if pause_sec:
            import time as _t; _t.sleep(pause_sec)

    if not frames:
        raise RuntimeError("No candle data returned. Try a shorter range or different product.")

    out = pd.concat(frames, axis=0).drop_duplicates(subset=["time"])
    out = out.set_index("time")[["open","high","low","close","volume"]]
    return out


end = datetime.utcnow()
start = end - timedelta(hours=24)

btc_1m = coinbase_candles_chunked("BTC-USD", granularity=60, start=start, end=end)
btc_1m.head()
```

```python
# compute returns and featueres

btc_1m["ret"] = np.log(btc_1m["close"] / btc_1m["close"].shift(1))
btc_1m["abs_ret"] = btc_1m["ret"].abs()
btc_1m = btc_1m.dropna()
```

```python
# intraday volume profile

btc_1m["minute"] = btc_1m.index.time
vol_by_min = btc_1m.groupby("minute")["volume"].mean()

plt.figure()
vol_by_min.plot()
plt.title("BTC-USD Avg Volume by Minute-of-Day")
plt.xlabel("Minute of Day"); plt.ylabel("Volume")
plt.show()
```

```python
# volume vs abs return aka does higher activity mean bigger moves

plt.figure()
plt.scatter(btc_1m["volume"], btc_1m["abs_ret"], alpha=0.4)
plt.title("BTC-SUD 1m: Volume vs |Return| (Coinbase)")
plt.xlabel("Volume"), plt.ylabel("|return|")
plt.show()

corr_v_vol = btc_1m[["volume","abs_ret"]].corr().iloc[0,1]
print("Correlation between volume and |return|: ", corr_v_vol)
```

```python
# market efficiency check (1-lag autocorr)

auto1 = btc_1m["ret"].autocorr(lag=1)
print("1-lag autocorrelation of 1m returns: ", auto1)
```
