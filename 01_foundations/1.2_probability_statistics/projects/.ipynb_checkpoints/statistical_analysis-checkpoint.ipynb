```python
# %pip install yfinance pandas numpy matplotlib scipy statsmodels

import warnings, math, itertools
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

import yfinance as yf
from scipy import stats
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.stats.diagnostic import acorr_ljungbox

plt.rcParams["figure.figsize"] = (9,5)
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
np.random.seed(42)
```

```python
def to_log_returns(px: pd.Series | pd.DataFrame) -> pd.DataFrame:
    r = np.log(px/px.shift(1)).dropna()
    return r.to_frame() if isinstance(r, pd.Series) else r

def var_es_empirical(r: pd.Series, alpha=0.01):
    """Left-tail VaR/ES for returns (losses are negative numbers)."""
    var = r.quantile(alpha)
    es = r[r <= var].mean()
    return float(var), float(es)

def var_es_parametric_normal(mu, sigma, alpha=0.01):
    z = stats.norm.ppf(alpha)
    var = mu + sigma * z
    # ES for normal (closed-form)
    es = mu - sigma * stats.norm.pdf(z) / alpha
    return float(var), float(es)

def var_es_parametric_t(df, loc, scale, alpha=0.01, mc=200000):
    # Use MC from fitted Student-t for ES
    x = stats.t.rvs(df, loc=loc, scale=scale, size=mc, random_state=42)
    q = np.quantile(x, alpha)
    es = x[x <= q].mean()
    return float(q), float(es)

def fit_student_t(r: pd.Series):
    df, loc, scale = stats.t.fit(r)
    return df, loc, scale

def moving_block_bootstrap_paths(R: pd.DataFrame, horizon=21, B=10000, block=10):
    """Block bootstrap portfolio returns over a multi-day horizon."""
    n = len(R)
    idx = np.arange(n)
    paths = []
    for _ in range(B):
        taken = []
        while len(taken) < horizon:
            start = np.random.randint(0, n - block)
            taken.extend(idx[start:start+block])
        taken = taken[:horizon]
        # sum log-returns ~ log of gross return; convert to simple return for horizon
        paths.append(R.iloc[taken].sum().to_frame().T)
    return pd.concat(paths, ignore_index=True)

def summarize_distribution_fit(r: pd.Series):
    jb = stats.jarque_bera(r)
    ad = stats.anderson(r, dist='norm')
    shapiro_stat, shapiro_p = stats.shapiro(r.sample(min(len(r), 5000), random_state=42))  # Shapiro cap
    skew = stats.skew(r, bias=False)
    kurt = stats.kurtosis(r, fisher=True, bias=False)
    return {
        "mean": r.mean(), "std": r.std(ddof=1),
        "skew": skew, "excess_kurtosis": kurt,
        "JB_stat": jb.statistic, "JB_p": jb.pvalue,
        "Shapiro_stat": shapiro_stat, "Shapiro_p": shapiro_p,
        "AD_stat": ad.statistic, "AD_crit": ad.critical_values
    }

def qqplot_series(r: pd.Series, title="Q-Q vs Normal"):
    plt.figure()
    stats.probplot(r, dist="norm", sparams=(r.mean(), r.std(ddof=1)), plot=plt)
    plt.title(title); plt.show()
```

```python
# daily log returns
R = np.log(raw / raw.shift(1)).dropna()
display(R.describe())
```

```python
# distribution analysis

r_spy = R["SPY"].dropna()
summary_spy = summarize_distribution_fit(r_spy)
display(summary_spy)
```

```python
# fit norma and student-t

mu_n, sigma_n = r_spy.mean(), r_spy.std(ddof=1)
df_t, loc_t, scale_t = stats.t.fit(r_spy)
print(f"Normal fit: mu={mu_n:.6g}, sigma={sigma_n:.6g}")
print(f"t fit: df={df_t:.2f}, loc={loc_t:.6g}, scale={scale_t:.6g}")
```

```python
# histogram + fitted PDFs
xs = np.linspace(r_spy.quantile(0.001), r_spy.quantile(0.999), 500)
plt.figure()
r_spy.plot(kind="hist", bins=120, density=True, alpha=0.6, label="SPY returns")
plt.plot(xs, stats.norm.pdf(xs, mu_n, sigma_n), label="Normal PDF")
plt.plot(xs, stats.t.pdf(xs, df_t, loc=loc_t, scale=scale_t), label=f"t(df={df_t:.1f}) PDF")
plt.title("SPY daily log returns — histogram & parametric fits")
plt.xlabel("return"); plt.ylabel("density"); plt.legend(); plt.show()
```

```python
# Q–Q plot vs Normal
plt.figure()
stats.probplot(r_spy, dist="norm", sparams=(mu_n, sigma_n), plot=plt)
plt.title("SPY daily log returns — Q–Q vs Normal"); plt.show()
```

```python
# tail risk value at risk and ES

rows = []
for a in (0.05, 0.01):
    var_emp, es_emp = var_es_empirical(r_spy, a)
    var_n, es_n     = var_es_parametric_normal(mu_n, sigma_n, a)
    var_t, es_t     = var_es_parametric_t(df_t, loc_t, scale_t, a)
    rows.append(pd.Series(
        {"VaR_emp":var_emp, "ES_emp":es_emp,
         "VaR_norm":var_n, "ES_norm":es_n,
         "VaR_t":var_t, "ES_t":es_t},
        name=f"alpha={a:.0%}"
    ))
display(pd.DataFrame(rows).T.round(6))
```

```python
# “3σ” exceedances check (under Normal)

thresh = mu_n - 3*sigma_n
observed = (r_spy < thresh).sum()
expected = len(r_spy) * stats.norm.cdf((thresh - mu_n)/sigma_n)
pd.Series({"threshold": thresh, "observed_count": observed, "normal_expected": expected}).round(4)
```

```python
# hypothesis testing

# stationary adf prices vs returns

def adf_report(series: pd.Series, name: str):
    stat, p, usedlag, nobs, crit, _ = adfuller(series.dropna(), autolag="AIC")
    return pd.Series({"series": name, "ADF_stat": stat, "p_value": p, "lags_used": usedlag, "n_obs": nobs})

adf_price = adf_report(raw["SPY"], "SPY price")
adf_rets  = adf_report(r_spy,       "SPY return")
display(pd.concat([adf_price, adf_rets], axis=1))
```

```python
# cross asset correlations
# we would not use correlated assets if we are trying to diversify a portfolio

from itertools import combinations

rows = []
for a,b in combinations(R.columns, 2):
    x, y = R[a].align(R[b], join="inner")
    pear_r, pear_p = stats.pearsonr(x, y)
    spear_r, spear_p = stats.spearmanr(x, y)
    rows.append({"pair": f"{a}-{b}",
                 "pearson_r": pear_r, "pearson_p": pear_p,
                 "spearman_r": spear_r, "spearman_p": spear_p})
corr_tests = pd.DataFrame(rows).sort_values("pearson_r", ascending=False)
display(corr_tests)
```

```python
# mean reversion / momentum

# Ljung–Box p-values at lags 1,5,10
lb = acorr_ljungbox(r_spy, lags=[1,5,10], return_df=True)
display(lb)
```

```python
# AR(1): r_t = alpha + beta*r_{t-1} + eps, with HAC (Newey–West) SEs
df_ar = pd.DataFrame({"r": r_spy})
df_ar["r_lag1"] = df_ar["r"].shift(1)
df_ar = df_ar.dropna()

# >>> IMPORTANT: pass a DataFrame so names are preserved
X = sm.add_constant(df_ar[["r_lag1"]])   # columns: ['const','r_lag1']
y = df_ar["r"]

ols = sm.OLS(y, X).fit()

n = len(df_ar)
L = int(np.floor(4 * (n/100)**(2/9)))    # HAC lag rule-of-thumb
hac = ols.get_robustcov_results(cov_type="HAC", maxlags=L)

# Wrap params/t/p into labeled Series using X column names
params = pd.Series(hac.params, index=X.columns)
tvals  = pd.Series(hac.tvalues, index=X.columns)
pvals  = pd.Series(hac.pvalues, index=X.columns)

out = pd.Series({
    "beta_lag1": params["r_lag1"],
    "t_HAC":     tvals["r_lag1"],
    "p_HAC":     pvals["r_lag1"],
    "HAC_lags":  L
}).round(6)

display(out)
```

```python
# monte carlo sim (port VaR/ES)

# Equal weights (edit if desired)
w = pd.Series(np.repeat(1/len(R.columns), len(R.columns)), index=R.columns, name="w")
r_p = R.dot(w)  # daily portfolio log return

pd.Series({
    "mean_daily": r_p.mean(),
    "std_daily": r_p.std(ddof=1),
    "ann_mean": 252 * r_p.mean(),
    "ann_vol": np.sqrt(252) * r_p.std(ddof=1)
}).round(6)
```

```python
mu_vec = R.mean().values
cov_mat = R.cov().values
df_mv = 6  # degrees of freedom for multivariate t

def simulate_portfolio(B=100000, horizon=1, dist="normal", seed=42):
    rng = np.random.default_rng(seed)
    if dist == "normal":
        sims = rng.multivariate_normal(mu_vec, cov_mat, size=(B, horizon))
    elif dist == "t":
        chi = rng.chisquare(df_mv, size=(B,1))
        scale = np.sqrt(df_mv/chi)  # (B,1)
        Z = rng.multivariate_normal(np.zeros(len(mu_vec)), cov_mat, size=(B, horizon))
        sims = (Z * scale[:,None,:]) + mu_vec  # add mean per day
    elif dist == "bootstrap":
        idx = rng.integers(0, len(R), size=(B, horizon))
        sims = R.values[idx, :]  # resample rows (iid)
    else:
        raise ValueError("dist must be {'normal','t','bootstrap'}")
    rp = sims @ w.values  # (B, horizon)
    return rp.sum(axis=1)  # sum log returns over horizon

def moving_block_bootstrap_paths(Rdf: pd.DataFrame, horizon=21, B=50000, block=10, seed=42):
    rng = np.random.default_rng(seed)
    n = len(Rdf)
    idx = np.arange(n)
    paths = np.empty(B)
    for b in range(B):
        taken = []
        while len(taken) < horizon:
            start = rng.integers(0, n - block)
            taken.extend(idx[start:start+block])
        taken = taken[:horizon]
        paths[b] = Rdf.iloc[taken].dot(w).sum()
    return paths
```

```python
# 1-day simulations
B = 100000
r1_norm = simulate_portfolio(B, 1, "normal")
r1_t    = simulate_portfolio(B, 1, "t")
r1_boot = simulate_portfolio(B, 1, "bootstrap")

# 21-day simulations
r21_norm  = simulate_portfolio(B, 21, "normal")
r21_t     = simulate_portfolio(B, 21, "t")
r21_block = moving_block_bootstrap_paths(R, horizon=21, B=50000, block=10)
```

```python
def var_es_from_samples(x, alpha=0.01):
    q = np.quantile(x, alpha)
    es = x[x <= q].mean()
    return float(q), float(es)

rows = []
for label, samps in [
    ("1d_Normal", r1_norm), ("1d_t", r1_t), ("1d_Bootstrap", r1_boot),
    ("21d_Normal", r21_norm), ("21d_t", r21_t), ("21d_Block", r21_block)
]:
    for a in (0.05, 0.01):
        var, es = var_es_from_samples(samps, a)
        rows.append({"horizon": label, "alpha": a, "VaR": var, "ES": es})

mc_table = pd.DataFrame(rows)
display(mc_table.pivot(index="horizon", columns="alpha", values=["VaR","ES"]).round(6))
```
